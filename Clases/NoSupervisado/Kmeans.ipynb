```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Crear datos, tendremos 300 datos con 2 caracter√≠sticas
Utilizaremos 6 centros

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=300, n_features=2, 
                           centers=6, cluster_std=1.8,random_state=101)
data[0].shape
```

```python
data[1]
```

Vemos los datos

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='brg')
```

Importamos Kmeans y definimos k=6

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6)
```

Aplicamos el algoritmo

```python
kmeans.fit(data[0])
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='brg')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='brg')
```
