#Arbol de clasificación con scikitlearn


Ahora veremos como implementar el árbol con sklearn



```
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```
iris= load_iris()
iris.target_names
```

```
x = iris.data[:,2:] #Largo y ancho de los pétalos
y = iris.target

arbolClasificador = DecisionTreeClassifier(max_depth=2)
arbolClasificador.fit(x,y)
```

```
#Este nos servirá para poder visualizar el árbol
from sklearn.tree import export_graphviz

export_graphviz(arbolClasificador, out_file= "arbolDeIris.dot",
                feature_names= iris.feature_names[2:],
                class_names= iris.target_names)

import graphviz

with open("arbolDeIris.dot") as archivoDot:
  dot_graph = archivoDot.read()
display(graphviz.Source(dot_graph))
```
